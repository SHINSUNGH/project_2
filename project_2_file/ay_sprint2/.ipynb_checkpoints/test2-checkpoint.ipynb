```python
df = pd.read_csv('./사용량, 의료인력수.csv')
```

```python
df = df.drop(columns='Unnamed: 0')
```

```python
df.info()
```

```python
df.columns
```

```python
df
```

```python
df['의료인력수'].unique()
```

```python
df = df.replace('-',0)
```

```python
df = df.astype({'의료인력수':'int'})
```

```python
df1 = df.groupby(['연도'])[['의료인력수']].sum().reset_index()
```

```python
df2 = df.groupby(['연도'])[['사용량']].sum().reset_index()
```

```python
df = pd.merge(df1,df2,on=['연도'])
```

```python
fig = make_subplots(specs=[[{"secondary_y": True}]])
fig.add_trace(go.Scatter(x=df['연도'], y=df['의료인력수'], mode='lines', name='의료인력수'), secondary_y=False)
fig.add_trace(go.Scatter(x=df['연도'], y=df['사용량'], mode='lines', name='전력사용량'), secondary_y=True)

fig.update_xaxes(tickmode='linear', dtick=1)
fig.update_yaxes(title_text="의료인력수", secondary_y=False)
fig.update_yaxes(title_text="전력사용량", secondary_y=True)

fig.update_layout(title='의료인력수 - 전력 사용량 추이')
fig.show()
```
