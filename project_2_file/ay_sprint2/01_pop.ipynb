```python
pop = pop.melt(id_vars=['연월'], value_vars=['서울특별시', '부산광역시', '대구광역시', '인천광역시', '광주광역시', '대전광역시',
       '울산광역시', '세종특별자치시', '경기도', '강원도', '충청북도', '충청남도', '전라북도', '전라남도',
       '경상북도', '경상남도', '제주특별자치도'], var_name='시도', value_name='인구수')
```

```python
pop
```

```python
pop = pop.astype({'연월':'str'})
```

```python
pop.info()
```

```python
year = []
month = []
for i in range(len(pop)):
    y = pop['연월'][i].split('.')[0]
    m = pop['연월'][i].split('.')[1]
    if m == '1':
        m = '10'
    year.append(y)
    month.append(m)    
pop['연도'] = year
pop['월'] = month
```
