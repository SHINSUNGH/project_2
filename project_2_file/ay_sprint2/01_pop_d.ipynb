```python
import pandas as pd
```

```python
pop_d = pd.read_csv("2011~2022_인구밀도.csv")
```

```python
date = pop_d['Unnamed: 0'].to_list()
```

```python
date
```

```python
pop_d['연월'] = date
```

```python
pop_d = pop_d.astype({'연월':'float'})
```

```python
pop_d = pop_d.melt(id_vars=['연월'], value_vars=['서울특별시', '부산광역시', '대구광역시', '인천광역시', '광주광역시', '대전광역시',
       '울산광역시', '세종특별자치시', '경기도', '강원도', '충청북도', '충청남도', '전라북도', '전라남도',
       '경상북도', '경상남도', '제주특별자치도'], var_name='시도', value_name='인구밀도')
```

```python
pop_d
```

```python
pop_d = pop_d.astype({'연월':'str'})
```

```python
year = []
month = []
for i in range(len(pop_d)):
    y = pop_d['연월'][i].split('.')[0]
    m = pop_d['연월'][i].split('.')[1]
    if m == '1':
        m = '10'
    year.append(y)
    month.append(m)    
pop_d['연도'] = year
pop_d['월'] = month
```
