```python
import pandas as pd
```

```python
df = pd.read_csv('./기간_관광객_전력소비량.csv')
```

```python
df =df.drop(columns='Unnamed: 0')
```

```python
df
```

```python
for i in df['기간']:
    i = i.split(' 월')[0]
```

```python
df['기간'] = df['기간'].astype('str')
df['기간'] = df['기간'].str.replace('.', '_')
```

```python
df['기간'] = df['기간'].str.replace(' 월', '')
```

```python
df
```

```python
year = []
month = []
for i in df['기간']:
    year.append(i.split('_')[0])
    month.append(i.split('_')[1])
year, month
```

```python
df['연도'] = year
df['월'] = month
```

```python
df = df.astype({'연도':'int','월':'int'})
```
