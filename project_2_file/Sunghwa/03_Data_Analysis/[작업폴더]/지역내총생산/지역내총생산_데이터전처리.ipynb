```python
import pandas as pd
```

<h3>1. 지역내총생산 원본데이터 전처리</h3>

```python
df1 = pd.read_csv("GRDP_시도__20230502171123.csv", encoding="cp949")
df1.head(3)
```

```python
# 0, 1 행과 '경제활동별(1)' 열은 필요하지 않으므로 제거
df2 = df1.drop(0)
df3 = df2.drop(1)
df4 = df3.drop(columns='경제활동별(1)')
# 2021년 column 이름의 p) 제거
df5 = df4.rename(columns={'2021 p)' : '2021'})
# 세종특별자치시의 2010~2012 결측치 0으로 처리
df6 = df5.fillna(0)
df7 = df6.reset_index()
df8 = df7.drop(columns='index')
df8.info()
```

```python
# 2010~2021 칼럼의 데이터타입 int64로 변경
df9 = df8.astype({'2010' : 'int64', '2011' : 'int64', '2012' : 'int64', '2013' : 'int64',
                  '2014' : 'int64', '2015' : 'int64', '2016' : 'int64', '2017' : 'int64',
                  '2018' : 'int64', '2019' : 'int64', '2020' : 'int64', '2021' : 'int64',})
df9.info()
```

```python
df9
```

```python
# 데이터를 세로로 이어붙이기(melt 사용)
df_melt = pd.melt(df9,
                    id_vars=['시도별(1)'],
                    value_vars=['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021'])
df_melt
```

```python
## column 이름 바꾸기
df_GRDP = df_melt.rename(columns={'시도별(1)' : '시도', 'variable' : '연도', 'value' : '지역내총생산'})
df_GRDP
```

```python
### csv 파일로 내보내기
df_GRDP.to_csv("시도_연도_GRDP.csv")
```
