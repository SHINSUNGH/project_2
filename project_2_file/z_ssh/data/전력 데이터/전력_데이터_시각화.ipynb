```python
df = pd.read_csv('전력 사용량 8차.csv')
```

```python
df.groupby(['연도'])
```

```python
df = df.groupby(['연도', '시도'])
```

```python
df
```

```python
df_set = df.drop(columns=['시도'])
```

```python
df_set.groupby(['연도'])
```

```python
dr_col = df_set.drop(columns=['연도'])
```

```python
plt.figure(figsize = (16,3))
plt.plot(dr_col, marker = 'o', markersize=1)
plt.grid(axis='y', linestyle='-', alpha=0.4)
plt.legend(dr_col.columns,
           loc='center left', bbox_to_anchor=(1, 0.5)) # 그래프 범례를 그래프 밖에 위치
plt.title('2010~2021 월별 사용량 (kwh)')
plt.xlabel('월')
plt.ylabel('사용량(kWh)')
# plt.autoscale(axis='x', tight=True)
plt.show()
```
