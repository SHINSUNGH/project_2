```python
import pandas as pd
```

```python
def col_change(df):
    df.columns = ['연도','시도','시군구','계약종별',
                  '1월','2월','3월','4월','5월',
                  '6월','7월','8월','9월','10월',
                  '11월','12월']

    df.drop([0, 1], axis=0, inplace=True)

    return df
```

```python
df2010 = pd.read_excel("시군구별 전력사용량(2010년).xlsx")
df2011 = pd.read_excel("시군구별 전력사용량(2011년).xlsx")
df2012 = pd.read_excel("시군구별 전력사용량(2012년).xlsx")
df2013 = pd.read_excel("시군구별 전력사용량(2013년).xlsx")
df2014 = pd.read_excel("150128 시군구별 전력사용량(2014년 1월~12월).xlsx")
df2015 = pd.read_excel("시군구별 전력사용량(2015년 12월).xlsx")
df2016 = pd.read_excel("시군구별 전력사용량(2016년 12월).xlsx")
df2017 = pd.read_excel("8.시군구별 전력사용량(홈페이지 게시용)_201712.xlsx")
df2018 = pd.read_excel("8.시군구별 전력사용량(홈페이지 게시용)_201812.xlsx")
df2019 = pd.read_excel("(2019)시군구별 전력사용량.xlsx")
df2020 = pd.read_excel("시군구별 전력사용량(홈페이지 게시용)_202012.xlsx")
df2021 = pd.read_excel("8.시군구별 전력사용량(홈페이지 게시용)_202112.xlsx")

df2010 = col_change(df2010).reset_index()
df2011 = col_change(df2011).reset_index()
df2012 = col_change(df2012).reset_index()
df2013 = col_change(df2013).reset_index()
df2014 = col_change(df2014).reset_index()
df2015 = col_change(df2015).reset_index()
df2016 = col_change(df2016).reset_index()
df2017 = col_change(df2017).reset_index()
df2018 = col_change(df2018).reset_index()
df2019 = col_change(df2019).reset_index()
df2020 = col_change(df2020).reset_index()
df2021 = col_change(df2021).reset_index()
```

```python
df_all = pd.concat([df2010,
                    df2011,
                    df2012,
                    df2013,
                    df2014,
                    df2015,
                    df2016,
                    df2017,
                    df2018,
                    df2019,
                    df2020,
                    df2021])

df_all
```
