```python
trace = go.Scatter(
        x=df_data['년월'],
        y=df_data['사용량'],
        mode='line',
        line=dict(
            color='blue'
        ),
        name='전력 사용량'
    )

# 레이아웃 생성
layout = go.Layout(
    title='전력 사용량',
    xaxis=dict(
        title='년월'
    ),
    yaxis=dict(
        title='사용량'
    )
)

# 레이아웃 조합 후 그래프 작성
fig = go.Figure(data=[trace], layout=layout)
fig
```
